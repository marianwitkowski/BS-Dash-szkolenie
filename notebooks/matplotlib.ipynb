```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kraje = ["USA", "Niemcy", "Rosja"]
m_brazowy = np.array([12, 10, 8])
m_srebrny = np.array([4, 2, 8])
m_zloty = [2, 2, 1]

plt.bar(kraje, m_brazowy, color="brown", width=0.5, label="B")
plt.bar(kraje, m_srebrny, color="silver", width=0.5, bottom=m_brazowy, label="S")
plt.bar(kraje, m_zloty, color="gold", width=0.5, bottom=m_brazowy+m_srebrny, label="Z")
plt.legend(ncol=2, loc="best")

plt.gcf().set_size_inches(11, 8)

#zapis do pliku graficznego
plt.savefig("wykres.jpg", dpi=300, quality=50)

plt.show()
```

```python
kraje = ["USA", "Niemcy", "Rosja"]
m_brazowy = np.array([12, 10, 8])
m_srebrny = np.array([4, 2, 8])
m_zloty =  np.array([2, 2, 1])

BW = .5

x1 = np.array([1,5,9])
plt.bar(x1, m_brazowy, color='brown', width=BW)

x2 = x1 + BW
plt.bar(x2, m_srebrny, color='silver', width=BW)

x3 = x2 + BW
plt.bar(x3, m_zloty, color='gold', width=BW)

plt.xticks(x2, kraje)
plt.show()
```

```python
# wykres kołowy
imiona = ["Jaś", "Małgosia", "Zuzia", "Tomek"]
zabawki = [ 10, 8, 18, 6 ]
kolory = ["blue", "cyan", "yellowgreen","#FF0033"]

plt.figure(figsize=(10,7))
plt.pie(zabawki, labels=imiona, autopct="%.1f %%", colors=kolory, shadow=True,
       explode=[0.1, 0, 0, .5], counterclock=True, startangle=123)
plt.show()
```
